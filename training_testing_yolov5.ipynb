## Post-training Quantization

After running the initial trained model on Android studio, from now onwards you can quantize the model to make its size even smaller

```python
# In the default optimization mode, both weights and activations are quantized to 8-bit integers.
import tensorflow as tf

# Loading the TensorFlow Lite model
interpreter = tf.lite.Interpreter(model_path="/content/drive/MyDrive/yolov5/yolov5/runs/train/exp6/weights/best-fp16.tflite")
interpreter.allocate_tensors()

# Defining a converter with post-training quantization options
converter = tf.lite.TFLiteConverter.from_saved_model("/content/drive/MyDrive/yolov5/yolov5/runs/train/exp6/weights/best_saved_model/")
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Convert the model to TensorFlow Lite with post-training quantization
quantized_tflite_model = converter.convert()

# Save the quantized model to a file
with open("quantized_model.tflite", "wb") as f:
    f.write(quantized_tflite_model)
```

## Quantization Aware Training

```python
!pip install torch torchvision
```

```python
import torch
from models.yolo import Model
from torch.quantization import fuse_modules

# Load and configure the model for QAT
model = Model(cfg="/content/drive/MyDrive/yolov5/yolov5/models/yolov5s.yaml", ch=3, nc=80)
model.load_state_dict(torch.load('/content/drive/MyDrive/yolov5/yolov5/yolov5s.pt'))

# Fuse the model modules
model = fuse_modules(model)

model.qconfig = torch.quantization.get_default_qat_qconfig('fbgemm')
torch.quantization.prepare_qat(model, inplace=True)
```
